### duplicates

```python
dup={
    "id":[1,2,3,4,1,2,3,4],
    "name":["sha","sow","kan","swa","sha","sow","kan","swa"]
}
df_dup=pd.DataFrame(dup)
df_dup
```

#### checking duplicates

```python
df_dup.duplicated(keep=False)
```

```python
df1_dup=df_dup[df_dup.duplicated(keep=False)]
df1_dup
```

#### dropping duplicates

```python
df1_dup.drop_duplicates()
```

#### replace the values

```python
dup=df1_dup.replace(to_replace=["sha","sow","kan","swa"],\
               value=["Sha","Sow","Kani","Swas"])
dup
```

```python
df_dup_drop=dup.drop_duplicates()
df_dup_drop
```
